```python
import pandas as pd
import numpy as np
import os
```

```python
hewlett = pd.read_csv('./dataset/hewlett/training_set_rel3.tsv', sep='\t', encoding='mac_roman')
hewlett.to_csv('./dataset/hewlett/train.csv')
```

```python
text_ids = []
texts = []
for i in os.listdir('./dataset/old/raw/'):
    text_ids.append(i.split('.')[0])
    with open(os.path.join('./dataset/old/raw/'+i)) as f:
        texts.append(f.read())

old = pd.DataFrame()
old['text_ids'] = text_ids
old['texts'] = texts
old.to_csv('./dataset/old/train.csv')
```
